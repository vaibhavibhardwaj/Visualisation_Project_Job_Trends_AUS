```python
 gender_stats3 = gender_stats[["Time","Employers, female (% of female employment) (modeled ILO estimate) [SL.EMP.MPYR.FE.ZS]",
                            "Employment in agriculture, female (% of female employment) (modeled ILO estimate) [SL.AGR.EMPL.FE.ZS]",
                             "Employment in industry, female (% of female employment) (modeled ILO estimate) [SL.IND.EMPL.FE.ZS]",
                             "Employment in services, female (% of female employment) (modeled ILO estimate) [SL.SRV.EMPL.FE.ZS]"]]
gender_stats3["Sex"]= "Female"
gender_stats3 = gender_stats3.drop(gender_stats3.index[0:31])
gender_stats3 = gender_stats3.reset_index()
gender_stats3 = gender_stats3.drop(["index"], axis = 1)
gender_stats3.head()
agf=gender_stats3[["Time","Sex","Employment in agriculture, female (% of female employment) (modeled ILO estimate) [SL.AGR.EMPL.FE.ZS]"]]
agf["Type"]= "agriculture"
agf.rename(columns = {"Employment in agriculture, female (% of female employment) (modeled ILO estimate) [SL.AGR.EMPL.FE.ZS]":"Percentage"},inplace = True)
agf

sf=gender_stats3[["Time","Sex","Employment in services, female (% of female employment) (modeled ILO estimate) [SL.SRV.EMPL.FE.ZS]"]]
sf["Type"]= "services"
sf.rename(columns = {"Employment in services, female (% of female employment) (modeled ILO estimate) [SL.SRV.EMPL.FE.ZS]":"Percentage"},inplace = True)
sf

inf=gender_stats3[["Time","Sex","Employment in industry, female (% of female employment) (modeled ILO estimate) [SL.IND.EMPL.FE.ZS]"]]
inf["Type"]= "industry"
inf.rename(columns = {"Employment in industry, female (% of female employment) (modeled ILO estimate) [SL.IND.EMPL.FE.ZS]":"Percentage"},inplace = True)
inf

emp_female = pd.concat([agf,sf,inf])
emp_female
```

```python
emp_mf = pd.concat([emp_female,emp_male])
emp_mf= emp_mf.reset_index()
emp_mf = emp_mf.drop("index",axis = 1)
emp_mf
emp_mf.to_csv (r'C:\Users\vaibh\Documents\expo_project\emp_mf.csv', index = False, header=True)
```

```python
job_stats = pd.read_csv("cf4b91ce-8b01-4ccf-8294-43d5cb2d2550_Data.csv") 
job_stat2 = pd.read_csv("746043e4-2172-4863-b6b3-430837affb80_Data.csv") #with series
job_stat2 = job_stat2.drop(["Country Name","Country Code"], axis = 1)
job_stat2 = job_stat2.drop([166,167,168,169,170])
job_stats = job_stats.drop(["Country Code","Country Name"], axis = 1)
job_stats = job_stats.drop([27,28,29,30,31])
job_stats = job_stats.replace(to_replace= np.nan,value = 0)
job_stat2.head()
#job_stats.to_csv (r'C:\Users\vaibh\Downloads\Data_Extract_From_Jobs (1)\job_stat.csv', index = False, header=True)
job_stats.head()
```

```python
jobs_statsm= job_stats[["Time",
                        "Unemployment with advanced education, male (% of male labor force with advanced education) [SL.UEM.ADVN.MA.ZS]",
                        "Unemployment with intermediate education, male (% of male labor force with intermediate education) [SL.UEM.INTM.MA.ZS]"]]

jobs_statsf = job_stats[["Time",
                        "Unemployment with advanced education, female (% of female labor force with advanced education) [SL.UEM.ADVN.FE.ZS]",
                        "Unemployment with intermediate education, female (% of female labor force with intermediate education) [SL.UEM.INTM.FE.ZS]"]]
jobs_statsm["Sex"]= "Male"
jobs_statsf["Sex"]= "Female"


am=jobs_statsm[["Time","Sex","Unemployment with advanced education, male (% of male labor force with advanced education) [SL.UEM.ADVN.MA.ZS]"]]
am["Education"]= "advanced"
am.rename(columns = {"Unemployment with advanced education, male (% of male labor force with advanced education) [SL.UEM.ADVN.MA.ZS]":"Percentage unemployment"},inplace = True)
am


in_m=jobs_statsm[["Time","Sex","Unemployment with intermediate education, male (% of male labor force with intermediate education) [SL.UEM.INTM.MA.ZS]"]]
in_m["Education"]= "intermediate"
in_m.rename(columns = {"Unemployment with intermediate education, male (% of male labor force with intermediate education) [SL.UEM.INTM.MA.ZS]":"Percentage unemployment"},inplace = True)
in_m


in_f=jobs_statsf[["Time","Sex","Unemployment with intermediate education, female (% of female labor force with intermediate education) [SL.UEM.INTM.FE.ZS]"]]
in_f["Education"]= "intermediate"
in_f.rename(columns = {"Unemployment with intermediate education, female (% of female labor force with intermediate education) [SL.UEM.INTM.FE.ZS]":"Percentage unemployment"},inplace = True)
in_f

af=jobs_statsf[["Time","Sex","Unemployment with advanced education, female (% of female labor force with advanced education) [SL.UEM.ADVN.FE.ZS]"]]
af["Education"]= "advanced"
af.rename(columns = {"Unemployment with advanced education, female (% of female labor force with advanced education) [SL.UEM.ADVN.FE.ZS]":"Percentage unemployment"},inplace = True)
af

edu_mf = pd.concat([af,in_f,in_m,am])
edu_mf

edu_mf.to_csv (r'C:\Users\vaibh\Documents\expo_project\edu_mf.csv', index = False, header=True)
```
